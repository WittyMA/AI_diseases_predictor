# COVID-19 Symptoms Prediction using Machine Learning

This notebook demonstrates the process of building a machine learning model to predict COVID-19 infection probability based on symptoms. It covers data loading, preprocessing, model training, and evaluation.

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Import data_preprocessing utility
from data_preprocessing import preprocess_tabular_data

# --- 1. Data Loading ---
data_path = 	'../data/raw/covid19_symptoms.csv'	 # Adjust path as necessary

try:
    data = pd.read_csv(data_path)
    print("Dataset loaded successfully.")
    print(data.head())
except FileNotFoundError:
    print(f"Error: {data_path} not found. Creating dummy data for demonstration.")
    # Create dummy data based on typical COVID-19 symptoms dataset structure
    np.random.seed(42)
    data = pd.DataFrame({
        	'Fever'	: np.random.choice([	'Yes'	, 	'No'	], 1000),
        	'Tiredness'	: np.random.choice([	'Yes'	, 	'No'	], 1000),
        	'Dry_Cough'	: np.random.choice([	'Yes'	, 	'No'	], 1000),
        	'Difficulty_in_Breathing'	: np.random.choice([	'Yes'	, 	'No'	], 1000),
        	'Sore_Throat'	: np.random.choice([	'Yes'	, 	'No'	], 1000),
        	'None_Sympton'	: np.random.choice([	'Yes'	, 	'No'	], 1000),
        	'Pains'	: np.random.choice([	'Yes'	, 	'No'	], 1000),
        	'Nasal_Congestion'	: np.random.choice([	'Yes'	, 	'No'	], 1000),
        	'Runny_Nose'	: np.random.choice([	'Yes'	, 	'No'	], 1000),
        	'Diarrhea'	: np.random.choice([	'Yes'	, 	'No'	], 1000),
        	'None_Experiencing'	: np.random.choice([	'Yes'	, 	'No'	], 1000),
        	'Age'	: np.random.randint(1, 100, 1000),
        	'Gender'	: np.random.choice([	'Male'	, 	'Female'	], 1000),
        	'Contact'	: np.random.choice([	'Yes'	, 	'No'	], 1000),
        	'Infection_Prob'	: np.random.choice([	'Yes'	, 	'No'	], 1000)
    })

# --- 2. Exploratory Data Analysis (EDA) ---
print("
Dataset Info:")
data.info()

print("
Dataset Description:")
print(data.describe(include=	'all'	))

print("
Missing values:")
print(data.isnull().sum())

# Target distribution
plt.figure(figsize=(6, 4))
data[	'Infection_Prob'	].value_counts().plot(kind=	'bar'	)
plt.title("COVID-19 Infection Probability Distribution")
plt.xlabel("Infection Probability")
plt.ylabel("Count")
plt.xticks(rotation=0)
plt.show()

# Symptom distribution
symptom_cols = [	'Fever'	, 	'Tiredness'	, 	'Dry_Cough'	, 	'Difficulty_in_Breathing'	, 	'Sore_Throat'	, 
                	'Pains'	, 	'Nasal_Congestion'	, 	'Runny_Nose'	, 	'Diarrhea'	]

fig, axes = plt.subplots(3, 3, figsize=(15, 12))
axes = axes.ravel()

for i, symptom in enumerate(symptom_cols):
    data[symptom].value_counts().plot(kind=	'bar'	, ax=axes[i])
    axes[i].set_title(f	'{symptom} Distribution'	)
    axes[i].set_xlabel(symptom)
    axes[i].set_ylabel(	'Count'	)
    axes[i].tick_params(axis=	'x'	, rotation=0)

plt.tight_layout()
plt.show()

# Age distribution by infection probability
plt.figure(figsize=(10, 6))
sns.boxplot(data=data, x=	'Infection_Prob'	, y=	'Age'	)
plt.title("Age Distribution by COVID-19 Infection Probability")
plt.show()

# --- 3. Data Preprocessing ---
# Define columns for preprocessing
numerical_cols = [	'Age'	]
categorical_cols = [
    	'Fever'	, 	'Tiredness'	, 	'Dry_Cough'	, 	'Difficulty_in_Breathing'	, 	'Sore_Throat'	, 	'None_Sympton'	,
    	'Pains'	, 	'Nasal_Congestion'	, 	'Runny_Nose'	, 	'Diarrhea'	, 	'None_Experiencing'	, 	'Gender'	, 	'Contact'	
]
target_column = 	'Infection_Prob'	

# Convert target to binary (0: No, 1: Yes)
data[target_column] = data[target_column].map({	'No'	: 0, 	'Yes'	: 1})

X_processed, y, preprocessor = preprocess_tabular_data(data, numerical_cols, categorical_cols, target_column)

print("
Shape of preprocessed features:", X_processed.shape)
print("First 5 rows of preprocessed features:")
print(X_processed.head())

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_processed, y, test_size=0.2, random_state=42, stratify=y)

print("Data split into training and testing sets.")
print(f"Training features shape: {X_train.shape}")
print(f"Testing features shape: {X_test.shape}")

# --- 4. Model Building (Random Forest) ---
model = RandomForestClassifier(n_estimators=100, random_state=42)

print("Training Random Forest model...")
model.fit(X_train, y_train)
print("Model training complete.")

# --- 5. Model Evaluation ---
y_pred = model.predict(X_test)
y_pred_proba = model.predict_proba(X_test)[:, 1]

accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)
roc_auc = roc_auc_score(y_test, y_pred_proba)

print("
Model Evaluation Metrics:")
print(f"Accuracy: {accuracy:.4f}")
print(f"Precision: {precision:.4f}")
print(f"Recall (Sensitivity): {recall:.4f}")
print(f"F1-Score: {f1:.4f}")
print(f"AUC-ROC: {roc_auc:.4f}")

# Classification Report
from sklearn.metrics import classification_report
print("
Classification Report:")
print(classification_report(y_test, y_pred))

# Confusion Matrix
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(6, 4))
sns.heatmap(cm, annot=True, fmt=	'd'	, cmap=	'Blues'	, 
            xticklabels=[	'No COVID-19'	, 	'COVID-19'	], yticklabels=[	'No COVID-19'	, 	'COVID-19'	])
plt.title("Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()

# ROC Curve
from sklearn.metrics import roc_curve
fpr, tpr, _ = roc_curve(y_test, y_pred_proba)
plt.figure(figsize=(6, 4))
plt.plot(fpr, tpr, color=	'darkorange'	, lw=2, label=	'ROC curve (area = %0.2f)'	 % roc_auc)
plt.plot([0, 1], [0, 1], color=	'navy'	, lw=2, linestyle=	'--'	)
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("Receiver Operating Characteristic (ROC) Curve")
plt.legend(loc=	'lower right'	)
plt.show()

# Feature Importance
feature_importance = pd.DataFrame({
    	'feature'	: X_processed.columns,
    	'importance'	: model.feature_importances_
}).sort_values(	'importance'	, ascending=False)

plt.figure(figsize=(12, 8))
sns.barplot(data=feature_importance.head(15), x=	'importance'	, y=	'feature'	)
plt.title("Top 15 Feature Importances")
plt.xlabel("Importance")
plt.show()

print("
Top 10 Most Important Features:")
print(feature_importance.head(10))

# --- 6. Save Model (Optional) ---
import joblib
joblib.dump(model, 	'covid19_symptoms_random_forest_model.pkl'	)
joblib.dump(preprocessor, 	'covid19_symptoms_preprocessor.pkl'	)
print("Model and preprocessor saved.")
```
