# Chest X-Ray COVID-19 Detection using Deep Learning

This notebook demonstrates the process of building a deep learning model to detect COVID-19 from chest X-ray images. It covers data loading, preprocessing, model training, and evaluation using Convolutional Neural Networks (CNNs).

```python
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout, GlobalAveragePooling2D
from tensorflow.keras.applications import ResNet50
from tensorflow.keras.optimizers import Adam
from sklearn.metrics import classification_report, confusion_matrix, roc_curve, auc
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os

# Import data_preprocessing utility
from data_preprocessing import create_image_data_generators

# --- 1. Data Setup ---
# Ensure you have downloaded the COVID-19 Radiography Database from Kaggle
# Download from: https://www.kaggle.com/datasets/tawsifurrahman/covid19-radiography-database
# Organize it in the following structure:
# data/chest_xray_covid19/
# ├── train/
# │   ├── COVID/
# │   ├── NORMAL/
# │   └── VIRAL_PNEUMONIA/ (optional)
# └── val/
#     ├── COVID/
#     ├── NORMAL/
#     └── VIRAL_PNEUMONIA/ (optional)

train_dir = 	'../data/chest_xray_covid19/train'	
val_dir = 	'../data/chest_xray_covid19/val'	

# Image parameters
IMG_HEIGHT = 224  # ResNet50 typically uses 224x224
IMG_WIDTH = 224
BATCH_SIZE = 16   # Smaller batch size for larger images
EPOCHS = 20

print(f"Training directory: {train_dir}")
print(f"Validation directory: {val_dir}")

# Check if directories exist
if not os.path.exists(train_dir):
    print(f"Warning: Training directory {train_dir} not found.")
    print("Creating dummy directory structure for demonstration.")
    os.makedirs(f"{train_dir}/COVID", exist_ok=True)
    os.makedirs(f"{train_dir}/NORMAL", exist_ok=True)
    os.makedirs(f"{train_dir}/VIRAL_PNEUMONIA", exist_ok=True)
    os.makedirs(f"{val_dir}/COVID", exist_ok=True)
    os.makedirs(f"{val_dir}/NORMAL", exist_ok=True)
    os.makedirs(f"{val_dir}/VIRAL_PNEUMONIA", exist_ok=True)
    
    # Create dummy image files
    import numpy as np
    from PIL import Image
    
    for i in range(5):
        dummy_img = Image.fromarray(np.random.randint(0, 255, (224, 224, 3), dtype=np.uint8))
        dummy_img.save(f"{train_dir}/COVID/dummy_{i}.jpg")
        dummy_img.save(f"{train_dir}/NORMAL/dummy_{i}.jpg")
        dummy_img.save(f"{train_dir}/VIRAL_PNEUMONIA/dummy_{i}.jpg")
        dummy_img.save(f"{val_dir}/COVID/dummy_{i}.jpg")
        dummy_img.save(f"{val_dir}/NORMAL/dummy_{i}.jpg")
        dummy_img.save(f"{val_dir}/VIRAL_PNEUMONIA/dummy_{i}.jpg")
    
    print("Dummy data created for demonstration.")

# --- 2. Data Loading and Preprocessing ---
# For multi-class classification (COVID, NORMAL, VIRAL_PNEUMONIA)
train_datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=15,
    width_shift_range=0.1,
    height_shift_range=0.1,
    shear_range=0.1,
    zoom_range=0.1,
    horizontal_flip=True,
    fill_mode=	'nearest'	
)

val_datagen = ImageDataGenerator(rescale=1./255)

train_generator = train_datagen.flow_from_directory(
    train_dir,
    target_size=(IMG_HEIGHT, IMG_WIDTH),
    batch_size=BATCH_SIZE,
    class_mode=	'categorical'	  # Multi-class classification
)

val_generator = val_datagen.flow_from_directory(
    val_dir,
    target_size=(IMG_HEIGHT, IMG_WIDTH),
    batch_size=BATCH_SIZE,
    class_mode=	'categorical'	
)

if train_generator and val_generator:
    print(f"Found {train_generator.samples} training images belonging to {train_generator.num_classes} classes.")
    print(f"Found {val_generator.samples} validation images belonging to {val_generator.num_classes} classes.")
    print(f"Class indices: {train_generator.class_indices}")
    num_classes = train_generator.num_classes
else:
    print("Error: Could not create data generators. Please check your data directory structure.")
    num_classes = 3  # Default for COVID, NORMAL, VIRAL_PNEUMONIA

# Visualize some sample images
def plot_sample_images(generator, num_images=9):
    fig, axes = plt.subplots(3, 3, figsize=(12, 12))
    axes = axes.ravel()
    
    # Get a batch of images and labels
    images, labels = next(generator)
    class_names = list(generator.class_indices.keys())
    
    for i in range(min(num_images, len(images))):
        axes[i].imshow(images[i])
        predicted_class = np.argmax(labels[i])
        axes[i].set_title(f"Class: {class_names[predicted_class]}")
        axes[i].axis(	'off'	)
    
    plt.tight_layout()
    plt.show()

if train_generator:
    plot_sample_images(train_generator)

# --- 3. Model Building ---
# Using Transfer Learning with ResNet50
def create_covid_detection_model(num_classes):
    base_model = ResNet50(weights=	'imagenet'	, include_top=False, input_shape=(IMG_HEIGHT, IMG_WIDTH, 3))
    
    # Freeze the base model initially
    base_model.trainable = False
    
    model = Sequential([
        base_model,
        GlobalAveragePooling2D(),
        Dense(256, activation=	'relu'	),
        Dropout(0.5),
        Dense(128, activation=	'relu'	),
        Dropout(0.5),
        Dense(num_classes, activation=	'softmax'	 if num_classes > 2 else 	'sigmoid'	)
    ])
    
    return model

model = create_covid_detection_model(num_classes)
print("Created ResNet50 Transfer Learning model for COVID-19 detection")

# Compile the model
if num_classes > 2:
    loss = 	'categorical_crossentropy'	
else:
    loss = 	'binary_crossentropy'	

model.compile(
    optimizer=Adam(learning_rate=0.0001),  # Lower learning rate for transfer learning
    loss=loss,
    metrics=[	'accuracy'	]
)

model.summary()

# --- 4. Model Training ---
if train_generator and val_generator:
    print("Starting model training...")
    
    # Add callbacks for better training
    from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau
    
    early_stopping = EarlyStopping(
        monitor=	'val_loss'	,
        patience=5,
        restore_best_weights=True
    )
    
    reduce_lr = ReduceLROnPlateau(
        monitor=	'val_loss'	,
        factor=0.2,
        patience=3,
        min_lr=0.00001
    )
    
    history = model.fit(
        train_generator,
        steps_per_epoch=train_generator.samples // BATCH_SIZE,
        epochs=EPOCHS,
        validation_data=val_generator,
        validation_steps=val_generator.samples // BATCH_SIZE,
        callbacks=[early_stopping, reduce_lr],
        verbose=1
    )
    
    print("Model training complete.")
    
    # Plot training history
    def plot_training_history(history):
        fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 4))
        
        # Plot accuracy
        ax1.plot(history.history[	'accuracy'	], label=	'Training Accuracy'	)
        ax1.plot(history.history[	'val_accuracy'	], label=	'Validation Accuracy'	)
        ax1.set_title(	'COVID-19 Detection Model Accuracy'	)
        ax1.set_xlabel(	'Epoch'	)
        ax1.set_ylabel(	'Accuracy'	)
        ax1.legend()
        ax1.grid(True)
        
        # Plot loss
        ax2.plot(history.history[	'loss'	], label=	'Training Loss'	)
        ax2.plot(history.history[	'val_loss'	], label=	'Validation Loss'	)
        ax2.set_title(	'COVID-19 Detection Model Loss'	)
        ax2.set_xlabel(	'Epoch'	)
        ax2.set_ylabel(	'Loss'	)
        ax2.legend()
        ax2.grid(True)
        
        plt.tight_layout()
        plt.show()
    
    plot_training_history(history)
    
    # --- 5. Model Evaluation ---
    print("Evaluating model...")
    
    # Evaluate on validation set
    loss, accuracy = model.evaluate(val_generator, verbose=0)
    print(f"Validation Loss: {loss:.4f}")
    print(f"Validation Accuracy: {accuracy:.4f}")
    
    # Get predictions
    val_generator.reset()
    predictions = model.predict(val_generator, verbose=0)
    
    if num_classes > 2:
        predicted_classes = np.argmax(predictions, axis=1)
        true_classes = val_generator.classes
    else:
        predicted_classes = (predictions > 0.5).astype(int).flatten()
        true_classes = val_generator.classes
    
    # Classification report
    class_names = list(val_generator.class_indices.keys())
    print("
Classification Report:")
    print(classification_report(true_classes, predicted_classes, target_names=class_names))
    
    # Confusion matrix
    cm = confusion_matrix(true_classes, predicted_classes)
    plt.figure(figsize=(8, 6))
    sns.heatmap(cm, annot=True, fmt=	'd'	, cmap=	'Blues'	, 
                xticklabels=class_names, yticklabels=class_names)
    plt.title("Confusion Matrix - COVID-19 Detection")
    plt.xlabel("Predicted")
    plt.ylabel("Actual")
    plt.show()
    
    # For binary classification or COVID vs others, compute ROC-AUC
    if num_classes == 2:
        fpr, tpr, _ = roc_curve(true_classes, predictions.flatten())
        roc_auc = auc(fpr, tpr)
        
        plt.figure(figsize=(6, 4))
        plt.plot(fpr, tpr, color=	'darkorange'	, lw=2, label=f	'ROC curve (AUC = {roc_auc:.2f})'	)
        plt.plot([0, 1], [0, 1], color=	'navy'	, lw=2, linestyle=	'--'	)
        plt.xlim([0.0, 1.0])
        plt.ylim([0.0, 1.05])
        plt.xlabel(	'False Positive Rate'	)
        plt.ylabel(	'True Positive Rate'	)
        plt.title(	'ROC Curve - COVID-19 Detection'	)
        plt.legend(loc=	'lower right'	)
        plt.grid(True)
        plt.show()
        
        print(f"ROC-AUC: {roc_auc:.4f}")
    
    # --- 6. Fine-tuning (Optional) ---
    print("
Optional: Fine-tuning the model...")
    print("Unfreezing the top layers of the base model for fine-tuning.")
    
    # Unfreeze the top layers of the base model
    base_model = model.layers[0]
    base_model.trainable = True
    
    # Fine-tune from this layer onwards
    fine_tune_at = 100
    
    # Freeze all the layers before the `fine_tune_at` layer
    for layer in base_model.layers[:fine_tune_at]:
        layer.trainable = False
    
    # Recompile with a lower learning rate
    model.compile(optimizer=Adam(learning_rate=0.00001/10),  # Even lower learning rate
                  loss=loss,
                  metrics=[	'accuracy'	])
    
    print(f"Fine-tuning from layer {fine_tune_at} onwards...")
    
    # Continue training for a few more epochs
    fine_tune_epochs = 5
    total_epochs = len(history.history[	'loss'	]) + fine_tune_epochs
    
    history_fine = model.fit(
        train_generator,
        steps_per_epoch=train_generator.samples // BATCH_SIZE,
        epochs=total_epochs,
        initial_epoch=len(history.history[	'loss'	]),
        validation_data=val_generator,
        validation_steps=val_generator.samples // BATCH_SIZE,
        verbose=1
    )
    
    print("Fine-tuning complete.")
    
    # Final evaluation
    final_loss, final_accuracy = model.evaluate(val_generator, verbose=0)
    print(f"Final Validation Loss: {final_loss:.4f}")
    print(f"Final Validation Accuracy: {final_accuracy:.4f}")
    
    # --- 7. Save Model ---
    model.save(	'covid19_detection_model.h5'	)
    print("Model saved as 'covid19_detection_model.h5'")
    
else:
    print("Cannot proceed with training due to data loading issues.")

# --- 8. Prediction Function ---
def predict_covid19(image_path, model, img_height=224, img_width=224):
    """
    Predicts COVID-19 from a chest X-ray image.
    
    Args:
        image_path (str): Path to the chest X-ray image
        model: Trained model
        img_height (int): Height to resize image to
        img_width (int): Width to resize image to
    
    Returns:
        tuple: (prediction_label, confidence_scores)
    """
    img = tf.keras.utils.load_img(image_path, target_size=(img_height, img_width))
    img_array = tf.keras.utils.img_to_array(img)
    img_array = tf.expand_dims(img_array, 0)  # Create a batch
    img_array = img_array / 255.0  # Normalize
    
    predictions = model.predict(img_array)[0]
    
    if len(predictions) > 2:  # Multi-class
        class_names = [	'COVID'	, 	'NORMAL'	, 	'VIRAL_PNEUMONIA'	]  # Adjust based on your classes
        predicted_class_idx = np.argmax(predictions)
        predicted_class = class_names[predicted_class_idx]
        confidence = predictions[predicted_class_idx]
        
        return predicted_class, dict(zip(class_names, predictions))
    else:  # Binary
        confidence = predictions[0]
        if confidence > 0.5:
            return "COVID-19", confidence
        else:
            return "NORMAL", 1 - confidence

# Example usage (uncomment to test with a real image):
# result, confidence = predict_covid19('/path/to/your/chest_xray.jpg', model)
# print(f"Prediction: {result}")
# print(f"Confidence scores: {confidence}")

print("
COVID-19 detection model training and evaluation complete!")
```
